```python
from builtins import enumerate, len
import pandas as pd
from ast import literal_eval
```

```python
data_src = "../data/data.csv"
df = pd.read_csv(data_src)
print("shape:",df.shape)
print("columns", df.columns)
```

```python
from nltk import word_tokenize
import re
REPL_NUMBER = "_REPL_NUMBER_"

word_counter = {}

def is_number_repl_isdigit(s):
    """ Returns True is string is a number. """
    return s.replace('.','',1).isdigit()

def tokenize(text):
    return word_tokenize(text)
    
def tokenize_doc(row):
    global word_counter
    text = row["review_headline"] + ". " + row["review_body"]
    text = str.lower(text)
    tokens = tokenize(text)
    tokens = list(filter(lambda x: re.match(r"\w(\w\d\-\_)*", x), tokens))
    
    tokens = list(map(lambda t: REPL_NUMBER if is_number_repl_isdigit(t) else t, tokens))
    
    for t in tokens:
        try: 
            word_counter[t] +=1
        except:
            word_counter[t] = 1
    
    return tokens
    
df["tokens"] = df.apply(tokenize_doc, axis=1)
```

```python
rare_words_threashold = 10
```

```python
rare_words = count_data[count_data <= rare_words_threashold].index.tolist()
print(len(rare_words))
```

```python
import sys


RARE_WORD = "_RARE_WORD_"
def replace_rare(inp):
    if type(inp) == list:
        tokens = inp
    else:
        tokens = literal_eval(inp)
    result = []
    for token in tokens:
        if token in rare_words:
            result.append(RARE_WORD)
        else:
            result.append(token)
    print (result)
    return result
    
df["tokens_rare"] = df.tokens.apply(replace_rare)
```
